```python
import pandas as pd
import os
file_path=os.path.join('.','data','raw')
full_path=os.path.join(file_path,'01_fund_master.csv')
file_read=pd.read_csv(full_path)
column=['fund_house','scheme_name','category','sub_category']
for a in column:
    file_read[a]=file_read[a].str.strip()
print("Trim the unwanted spaces")
column1=['risk_category','benchmark']
for a in column1:
    print(file_read[a].value_counts())
new_file_path=os.path.join('.','data','processed')
new_file_full_path=os.path.join(new_file_path,'clean_fund_master.csv')
file_read.to_csv(new_file_full_path,index=False)
print("Cleaned fund master file")
```

```python
import pandas as pd
import os
file_path=os.path.join('.','data','raw')
full_path=os.path.join(file_path,'03_aum_by_fund_house.csv')
file_read=pd.read_csv(full_path)
file_read['date']=pd.to_datetime(file_read['date'],errors='coerce')
file_read['fund_house']=file_read['fund_house'].str.strip()
file_read['aum_crore']=pd.to_numeric(file_read['aum_crore'],errors='coerce')
file_read['aum_crore']=file_read['aum_crore'].fillna(0)
new_file_path=os.path.join('.','data','processed')
new_file_full_path=os.path.join(new_file_path,'clean_aum_by_fund_master.csv')
file_read.to_csv(new_file_full_path,index=False)
print("Cleaned aum by fund master file")
```

```python
import pandas as pd
import os
file_path=os.path.join('.','data','raw')
full_path=os.path.join(file_path,'09_portfolio_holdings.csv')
file=pd.read_csv(full_path)
file_read['portfolio_date']=pd.to_datetime(file_read['portfolio_date'],errors='coerce')
column=['weight_pct','market_value_cr','current_price_inr']
for a in column:
    file[a]=pd.to_numeric(file[a],errors='coerce')
    file[a]=file[a].fillna(0)
new_file_path=os.path.join('.','data','processed')
new_file_full_path=os.path.join(new_file_path,'clean_portfolio_holdings.csv')
file_read.to_csv(new_file_full_path,index=False)
print("Cleaned portfolio_holdingsfile")
```
