```python
import os
import requests
import pandas as pd

path = './data/raw'
nav_schemes = {
    "hdfc": "125497",
    "SBI_Bluechip": "119551",
    "ICICI_Bluechip": "120503",
    "Nippon_Large_Cap": "118632",
    "Axis_Bluechip": "119092",
    "Kotak_Bluechip": "120841"
}
for name, code in nav_schemes.items():
    try:
        api_url = f"https://api.mfapi.in/mf/{code}"
        json_data = requests.get(api_url).json()
        print(f"\n---Scheme name: {name} (Code: {code}) ---")
        print("Keys:", json_data.keys())      
        print("First Row:", json_data['data'][0])  
        nav_list = json_data['data']
        df_scheme = pd.DataFrame(nav_list)
        file_name = f"live_{name}_nav.csv"
        full_path = os.path.join(path, file_name)
        df_scheme.to_csv(full_path, index=False)
        print(f"Saved successfully at: {full_path}")
        display(df_scheme.head(2))

    except Exception as e:
        print(e)
```

```python
import os
import pandas as pd
path = "./data/raw/01_fund_master.csv"
try:
    df_master = pd.read_csv(path)
    print("\n---VALIDATING AMFI CODES ---")
    master_codes = set(df_master["amfi_code"].unique())
    print(f" Master File  {len(master_codes)}")
    fetched_codes = {125497, 119551, 120503, 118632, 119092, 120841}
    validated_count = len(master_codes & fetched_codes)
    missing_codes = master_codes - fetched_codes
    print(f"Successfully Validated Codes: {validated_count}")
    print(f"Unfetched Codes: {len(missing_codes)}")
except FileNotFoundError as e:
    print(e)
```
